### Perform exploratory data analysis on the netflix_data.csv data to understand more about movies from the 1990s decade

```python
import pandas as pd
```

```python
df = pd.read_csv('netflix_data.csv')
```

```python
# Getting acquainted with the data in the table
df.head()
```

```python
# Gettng an overview of our data
# We can understand that this table has 11 columns, 4812 rows and there are not null values
df.info()
```

#### 1) What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

```python
# using filters to see all films with release year between 1990 and 1999
df[(df['release_year'] < 2000) & (df['release_year'] >= 1990)]
```

```python
# using value_counts() to count the values of all durations
df[(df['release_year'] < 2000) & (df['release_year'] >= 1990)]['duration'].value_counts()
```

```python
# getting the most frequent movie duration. We can see that there 7 films with 94 min duration which were released
# between 1990 and 1999
duration = df[(df['release_year'] < 2000) & (df['release_year'] >= 1990)]['duration'].value_counts().head(1)
print(duration)
```

```python
# getting the most frequent movie duration
duration = df[(df['release_year'] < 2000) & (df['release_year'] >= 1990)]['duration'].value_counts().head(1).index[0]
print(duration)
```

#### 2) A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
# used filters to specify our table for our needs. 
df[(df["duration"] < 90) & (df['genre'] == 'Action') & (df['release_year'] < 2000) & (df['release_year'] >= 1990)]
```

```python
# filtered by duration column
short_movie_count = df[(df["duration"] < 90) & (df['genre'] == 'Action') & (df['release_year'] < 2000) & (df['release_year'] >= 1990)]['duration']
print(short_movie_count)
```

```python
# got the number of films
short_movie_count = df[(df["duration"] < 90) & (df['genre'] == 'Action') & (df['release_year'] < 2000) & (df['release_year'] >= 1990)]['duration'].count()
print(short_movie_count)
```
